# Data merging: October and November data

#### We split up both csv files into a product and customer one for both database tables.
#### The filter out redundant product data and only keep one entry per product (unique 'product_id').

```python
import pandas as pd
```

```python
data_Oct = pd.read_csv("C:/Users/timma/OneDrive - Handelshögskolan i Stockholm/HSG/Big Data Analytics/project/Oct19.csv")
data_customer_Oct = data_Oct.iloc[:,[0,1,2,7,8]]
data_product_Oct = data_Oct.iloc[:,2:7].groupby('product_id').first()
```

```python
data_Nov = pd.read_csv("C:/Users/timma/OneDrive - Handelshögskolan i Stockholm/HSG/Big Data Analytics/project/Nov19.csv")
data_customer_Nov = data_Nov.iloc[:,[0,1,2,7,8]]
data_product_Nov = data_Nov.iloc[:,2:7]
```

```python
data_10k = pd.read_csv("C:/Users/timma/OneDrive - Handelshögskolan i Stockholm/HSG/Big Data Analytics/project/data_10k_rows.csv")
data_10k_customer = data_10k.iloc[:,[0,1,2,7,8]]
data_10k_product = data_10k.iloc[:,2:7]
```
